```python
#Import the following packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the data
data = pd.read_csv('MySensorData.csv')
data.head()
data.info()
```

```python
#Plot the data with respect to time for all the three axes
plt.figure(figsize = (15,15))

plt.subplot(3,1,1)
plt.plot(data['Time (s)'],data['Linear Acceleration x (m/s^2)'])
plt.grid()
plt.title('Acceleration x')
plt.xlabel('Time')
plt.ylabel('a_x')
plt.axis([0,30,-10,10])

plt.subplot(3,1,2)
plt.plot(data['Time (s)'],data['Linear Acceleration y (m/s^2)'])
plt.grid()
plt.title('Acceleration y')
plt.xlabel('Time')
plt.ylabel('a_y')
plt.axis([0,30,-10,10])

plt.subplot(3,1,3)
plt.plot(data['Time (s)'],data['Linear Acceleration z (m/s^2)'])
plt.grid()
plt.title('Acceleration z')
plt.xlabel('Time')
plt.ylabel('a_z')
plt.axis([0,30,-10,10])
plt.show()
```
